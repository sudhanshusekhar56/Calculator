```python
from tkinter import *

#tkinter window
root=Tk()
root.geometry("280x410+300+300")
root.resizable(0,0)
root.title("Calculator")

#declaring variables
val=""
operator=""
data=StringVar()

#numeric button's function
def btn1():
    global val
    val=val+"1"
    data.set(val)
    
def btn2():
    global val
    val=val+"2"
    data.set(val)
    
def btn3():
    global val
    val=val+"3"
    data.set(val)

def btn4():
    global val
    val=val+"4"
    data.set(val)
    
def btn5():
    global val
    val=val+"5"
    data.set(val)
    
def btn6():
    global val
    val=val+"6"
    data.set(val)
    
def btn7():
    global val
    val=val+"7"
    data.set(val)
    
def btn8():
    global val
    val=val+"8"
    data.set(val)
    
def btn9():
    global val
    val=val+"9"
    data.set(val)
    
def btn0():
    global val
    val=val+"0"
    data.set(val)
    
#operator functions
def btnadd():
    global val
    global temp
    operator="+"
    val=val+"+"
    data.set(val)
    
def btnsub():
    global val
    global temp
    operator="-"
    val=val+"-"
    data.set(val)
    
def btnmult():
    global val
    global temp
    operator="*"
    val=val+"*"
    data.set(val)
    
def btndiv():
    global val
    global temp
    operator="/"
    val=val+"/"
    data.set(val)

#cancel button function
def btnc():
    global val
    global temp
    global operator
    operator=""
    val=""
    data.set(val)
    
#equal button function
def result():
    global val
    global temp
    global operator
    c=str(eval(val))
    operator=""
    data.set(c)

    
    

#creating top label
lb=Label(root,anchor=SE,font=("verdana",19),textvariable=data,bg="white",fg="black")
lb.pack(expand=TRUE,fill="both")

#creating frames for buttons
row1=Frame(root)
row1.pack(expand=TRUE,fill="both")
row2=Frame(root)
row2.pack(expand=TRUE,fill="both")
row3=Frame(root)
row3.pack(expand=TRUE,fill="both")
row4=Frame(root)
row4.pack(expand=TRUE,fill="both")


#creating buttons for calc
b1=Button(row1,text="1",font=("verdana",18),relief=GROOVE,border=0,command=btn1)
b1.pack(side=LEFT,expand=TRUE,fill="both")

b2=Button(row1,text="2",font=("verdana",18),relief=GROOVE,border=0,command=btn2)
b2.pack(side=LEFT,expand=TRUE,fill="both")

b3=Button(row1,text="3",font=("verdana",18),relief=GROOVE,border=0,command=btn3)
b3.pack(side=LEFT,expand=TRUE,fill="both")

b_add=Button(row1,text="+",font=("verdana",18),relief=GROOVE,border=0,command=btnadd)
b_add.pack(side=LEFT,expand=TRUE,fill="both")



b4=Button(row2,text="4",font=("verdana",18),relief=GROOVE,border=0,command=btn4)
b4.pack(side=LEFT,expand=TRUE,fill="both")

b5=Button(row2,text="5",font=("verdana",18),relief=GROOVE,border=0,command=btn5)
b5.pack(side=LEFT,expand=TRUE,fill="both")

b6=Button(row2,text="6",font=("verdana",18),relief=GROOVE,border=0,command=btn6)
b6.pack(side=LEFT,expand=TRUE,fill="both")

b_sub=Button(row2,text="-",font=("verdana",18),relief=GROOVE,border=0,command=btnsub)
b_sub.pack(side=LEFT,expand=TRUE,fill="both")



b7=Button(row3,text="7",font=("verdana",18),relief=GROOVE,border=0,command=btn7)
b7.pack(side=LEFT,expand=TRUE,fill="both")

b8=Button(row3,text="8",font=("verdana",18),relief=GROOVE,border=0,command=btn8)
b8.pack(side=LEFT,expand=TRUE,fill="both")

b9=Button(row3,text="9",font=("verdana",18),relief=GROOVE,border=0,command=btn9)
b9.pack(side=LEFT,expand=TRUE,fill="both")

b_mult=Button(row3,text="*",font=("verdana",18),relief=GROOVE,border=0,command=btnmult)
b_mult.pack(side=LEFT,expand=TRUE,fill="both")



b_c=Button(row4,text="C",font=("verdana",18),relief=GROOVE,border=0,command=btnc)
b_c.pack(side=LEFT,expand=TRUE,fill="both")

b0=Button(row4,text="0",font=("verdana",18),relief=GROOVE,border=0,command=btn0)
b0.pack(side=LEFT,expand=TRUE,fill="both")

b_equal=Button(row4,text="=",font=("verdana",18),relief=GROOVE,border=0,command=result)
b_equal.pack(side=LEFT,expand=TRUE,fill="both")

b_div=Button(row4,text="/",font=("verdana",18),relief=GROOVE,border=0,command=btndiv)
b_div.pack(side=LEFT,expand=TRUE,fill="both")


root.mainloop()
```
